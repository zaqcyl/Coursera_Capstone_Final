# Property in New York City

__by Zach Cylinder__

__Objective__

My final project will be analyzing property values in New York City. I am choosing this since it is currently and will become a useful project for me as I begin my journey to move to New York City. With this project, I am seeking to define and segment criteria for various tiers of housing across the five boroughs of the city. I have found different maps, but not one that is comprehensive nor one that fits my needs.

__Data__

I will be using data from the City of New York obtained from kaggle.com. The dataset is called "NYC Property Sales" and details all of the properties sold in New York City from September 2016 to September 2017. While property sales do fluctuate – especially this year due to COVID-19 – I am seeking to spot trends and figure out which areas of the city are more affordable.
